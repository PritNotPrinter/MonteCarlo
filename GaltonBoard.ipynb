### Galton Boards

```python
import random 
import statistics
endContainer = []
count = 0
numTests = 10000000

def galton(numRows): # Takes parameter of number of rows to have
    rightTurns = 0 # Stores how many right bounces the ball went
    for row in range(numRows): # Iterates over the number of rows
        rightTurns += random.choice([0, 1]) # For each row, add 0 (left) or 1 (right) to the sum
    return rightTurns # Returns the var

while count < numTests:
    endContainer.append(galton(5))
    count += 1
    
print(endContainer)
print(statistics.mean(endContainer))

```

# Trying to output to CSV file

# Adding the Histogram (& maybe boxplot) to Galton

```python
# Imports
import random 
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Variables

endContainer = []
count = 0
numTests = 100
numRows = 5

# Galton Board function

def galton(numRows): # Takes parameter of number of rows to have
    rightTurns = 0 # Stores how many right bounces the ball went
    for row in range(numRows): # Iterates over the number of rows
        rightTurns += random.choice([0, 1]) # For each row, add 0 (left) or 1 (right) to the sum
    return rightTurns # Returns the var

# Main loop to simulate many balls

while count < numTests:
    endContainer.append(galton(numRows)) # Adds the result per ball from Galton to the list
    count += 1

# Printing and returning data
print(statistics.mean(endContainer)) # Prints the average
dfEndContainer = pd.DataFrame(endContainer) # turning the list into a dataframe to turn into csv
dfEndContainer.to_csv("galtonOutput.csv", index = False) # Turns the dataframed list to a csv file, ignoring the index (the column)

# Loading Galton simulation data from CSV
path = "galtonOutput.csv"
galtonData = pd.read_csv(path)

# Creating Histogram
sns.set(style="dark")
plt.figure(figsize=[10,6])
sns.histplot(galtonData.iloc[:, 0], color="rebeccapurple", bins = numRows+1) # Uses data from the first column of the CSV, rather than a specific name of column. bins > numRows because of how 0-start indexing works
plt.title("Distribution of Balls in Galton Model")
plt.xlabel("Container number (net right turns)")
plt.ylabel("Number of Balls in Container")
plt.show()
```
